#  TALLER RSA
1. ¿Cuál es el problema que resuelve RSA, y por qué es necesario introducir un algoritmo de este tipo?\
El RSA ayuda a dar seguridad a la hora de compartir información entre dos  usuarios de tal manera que al enviar un mensaje el remitente lo incripta y solo el destinatario puede desencriptar el mensaje.



2. Desarrollo matemático del algoritmo. Se deben especificar todos los pasos del algoritmo y se debe explicar de forma detallada cada uno de los pasos. Si el algoritmo hace uso de algunos resultados formales (teoremas) estos deben enunciarse y probarse de ser necesario.\
El algorimo de RSA:
  1. Escoger dos numeros primos distintos p y q.Cada uno de estos se escoge de manera random y se prueba que cada uno de estos es primo usando el pequeño teorema de Fermat\
   El teorema de fermat es:\
    sea $x$ un primo, entonces $a^x = a (mod x)$
  2. Se calcula $n = p \times q $
  3. Se calcula la función de euler
  4. Se escoge un entero positivo $e$ menor que $\phi (n)$, que sea coprimo con $\phi (n)$,para calcular esto se usa el algoritmo de euclides donde el mcd($\phi (n)$, e) debe ser 1\
  Algoritmo de euclides para calcular mcd(a,b):

    1. Divir el numero más grande(por ejemplo a) por el numero más pequeño\
    $a = q_1* b + r_1$
    2. Tomar b como el numero mas grande y se divide entre el residuo hallado en el paso anterior\
    $ b = q_2 * r_1 + r_2 $
    3. Tomar el residuo mas grande del paso anterior y se divide entre el más pequeño\
    $r_1 = q_3 * r_2 + r_3
    4. Repetir el paso el paso anterior hasta que el nuevo residuo obtenido sea 0\
    $r_n = q_{n+2} * r_{n+1} + 0$
    5. Tomar a $r_{n+1}$ como el mcd(a,b)
  5. Se determina $d$ ,el inverso multimplicativo de $e$ $(mod$ $n)$, para esto se usa el teorema extendido de euclides\
    Teorema de euclides extendido:\
    Suponga a y b enteros, no cero.Entonces existen  enteros m y n tales que\
     mcd(a,b) = m*a + n*b\
     \
    Para que este teorema de euclides extendido se pueda dar se realiza el algoritmo de euclides extendido\
    Algoritmo extendido de euclides para calcular (a,b) = m*a + n*b:
      1. Realizar el algoritmo de euclides
      2. Tomar $r_n$ y expresarlo de la ultima manera que se expreso
      3. Expresar cada uno de los residuos que hay en la ecuación del paso anterior
      4. Expresar los residuos de la ecuación del paso anterior 
      5. Repetir la ecuación hasta expresar $r_1$
      6. Resolver la ecuación en terminos de $a$ y $b$\
    En este caso se busca que mcd($\phi (n)$,e) = m*$\phi (n)$ + n*e = 1
    donde n sera el inverso multiplicativo de e (mod$\phi (n)$, es decir d) 
  6. Determinar la clave publica como $(n,e)$
  7. Determinar la clave privada como $(n,d)$
  8. Pasar cada letra del mensaje a codigo ASCII
  9. Cifrar cada una de las letras con la clave publica usando la siguiente formula:\ 
    $c = m^{e} (mod n)$\ 
    donde $m$ es la letra expresado en ASCII , $e$ es la segunda parte de la clave publica, $n$ la primera parte de la clave publica.
  10. Poner cada una de las letras cifradas en una string.
  11. Decifrar el cada uno de los caracteres del string del paso anterior con la clave privada usando la siguiente formula :\
    $m = c^{d} (mod n)$\
    donde $m$ es el caracter expresado en ASCII, $d$ es la segunda parte de la clave privada, $n$ la primera parte de la clave publica.

#ALGORITMO DE LUCAS


```
from math import *


def n_lucas(n):
    if n == 0:
        return 2
    if n == 1:
        return 1
    return int(pow((1 + sqrt(5))/2, n) + pow((1 - sqrt(5))/2, n))


def lucas_grande(n):
    lu = 0
    i = 0
    if n == 0:
        return 1
    while lu <= n:
        lu = int(pow((1 + sqrt(5))/2, i) + pow((1 - sqrt(5))/2, i))
        i = i + 1
    return int(lu)


def lucas_cercano(n):
    if n == 0:
        return 1
    antes = 0
    despues = 0
    lu = 0
    i = 0
    while lu <= n:
        antes = lu
        lu = int(pow((1 + sqrt(5))/2, i) + pow((1 - sqrt(5))/2, i))
        despues = lu
        i = i + 1
    anterior = n - antes
    siguiente = despues - n
    if anterior <= siguiente:
        lu = antes
    else:
        lu = despues
    return lu


print("el 18 th número de Lucas es:")
print(n_lucas(18))
print("el número de Lucas más grade que 100:")
print(lucas_grande(100))
print("el número de Lucas más cercano a 1000:")
print(lucas_cercano(1000))
```
